```python
from data.dataset import dataloaders, class_names
from models.resnet import get_resnet
from models.vit import get_vit
from utils.training import train_model
from utils.visualization import test_and_explain
import torch
import torch.optim as optim
from torch.nn import CrossEntropyLoss
from config import NUM_EPOCHS_RESNET, NUM_EPOCHS_VIT, LEARNING_RATE, DEVICE
```

```python
# Entraînement du ResNet
model_resnet = get_resnet()
criterion = CrossEntropyLoss()
optimizer = optim.Adam(model_resnet.fc.parameters(), lr=LEARNING_RATE)
model_resnet, history_resnet = train_model(model_resnet, dataloaders, criterion, optimizer, NUM_EPOCHS_RESNET, class_names)
torch.save(model_resnet.state_dict(), 'resnet18_fake_faces.pth')
```

```python
# Entraînement du ViT
model_vit = get_vit()
optimizer_vit = optim.Adam(model_vit.heads.head.parameters(), lr=LEARNING_RATE)
model_vit, history_vit = train_model(model_vit, dataloaders, criterion, optimizer_vit, NUM_EPOCHS_VIT)
torch.save(model_vit.state_dict(), 'vit_fake_faces.pth')
```

```python
# Tests et explications
test_and_explain(model_resnet, r'rvf10k/valid/fake/0AIFZB4IE6.jpg', true_label='fake')
test_and_explain(model_resnet, r'rvf10k/valid/real/00042.jpg', true_label='real')
test_and_explain(model_vit, r'rvf10k/valid/fake/0AIFZB4IE6.jpg', true_label='fake', method='vit')
test_and_explain(model_vit, r'rvf10k/valid/real/00042.jpg', true_label='real', method='vit')
```

```python
# Visualisation et comparaison des résultats
import matplotlib.pyplot as plt
import numpy as np

# Conversion des accuracies en pourcentage si nécessaire
def convert_acc(acc_list):
    """Convertit une liste d'accuracy en pourcentage si elle est en format décimal"""
    if len(acc_list) > 0 and max(acc_list) <= 1.0:
        return [x * 100 for x in acc_list]
    return acc_list

history_resnet['train_acc'] = convert_acc(history_resnet['train_acc'])
history_resnet['val_acc'] = convert_acc(history_resnet['val_acc'])
history_vit['train_acc'] = convert_acc(history_vit['train_acc'])
history_vit['val_acc'] = convert_acc(history_vit['val_acc'])

fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# ===== 1. Loss ResNet =====
epochs_resnet = range(1, len(history_resnet['train_loss']) + 1)
axes[0, 0].plot(epochs_resnet, history_resnet['train_loss'], 'b-', label='Train', linewidth=2)
axes[0, 0].plot(epochs_resnet, history_resnet['val_loss'], 'r-', label='Validation', linewidth=2)
axes[0, 0].set_xlabel('Époque', fontsize=11)
axes[0, 0].set_ylabel('Loss', fontsize=11)
axes[0, 0].set_title('ResNet18 - Évolution de la Loss', fontsize=12, fontweight='bold')
axes[0, 0].legend(fontsize=10)
axes[0, 0].grid(True, alpha=0.3)

# ===== 2. Loss ViT =====
epochs_vit = range(1, len(history_vit['train_loss']) + 1)
axes[0, 1].plot(epochs_vit, history_vit['train_loss'], 'g-', label='Train', linewidth=2)
axes[0, 1].plot(epochs_vit, history_vit['val_loss'], 'orange', label='Validation', linewidth=2)
axes[0, 1].set_xlabel('Époque', fontsize=11)
axes[0, 1].set_ylabel('Loss', fontsize=11)
axes[0, 1].set_title('Vision Transformer - Évolution de la Loss', fontsize=12, fontweight='bold')
axes[0, 1].legend(fontsize=10)
axes[0, 1].grid(True, alpha=0.3)

# ===== 3. Comparaison Loss Validation =====
axes[0, 2].plot(epochs_resnet, history_resnet['val_loss'], 'b-', marker='o', label='ResNet18', linewidth=2)
axes[0, 2].plot(epochs_vit, history_vit['val_loss'], 'g-', marker='s', label='Vision Transformer', linewidth=2)
axes[0, 2].set_xlabel('Époque', fontsize=11)
axes[0, 2].set_ylabel('Loss Validation', fontsize=11)
axes[0, 2].set_title('Comparaison - Loss Validation', fontsize=12, fontweight='bold')
axes[0, 2].legend(fontsize=10)
axes[0, 2].grid(True, alpha=0.3)

# ===== 4. Accuracy ResNet =====
axes[1, 0].plot(epochs_resnet, history_resnet['train_acc'], 'b-', label='Train', linewidth=2)
axes[1, 0].plot(epochs_resnet, history_resnet['val_acc'], 'r-', label='Validation', linewidth=2)
axes[1, 0].set_xlabel('Époque', fontsize=11)
axes[1, 0].set_ylabel('Accuracy (%)', fontsize=11)
axes[1, 0].set_title('ResNet18 - Évolution de l\'Accuracy', fontsize=12, fontweight='bold')
axes[1, 0].legend(fontsize=10)
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].set_ylim([0, 105])

# ===== 5. Accuracy ViT =====
axes[1, 1].plot(epochs_vit, history_vit['train_acc'], 'g-', label='Train', linewidth=2)
axes[1, 1].plot(epochs_vit, history_vit['val_acc'], 'orange', label='Validation', linewidth=2)
axes[1, 1].set_xlabel('Époque', fontsize=11)
axes[1, 1].set_ylabel('Accuracy (%)', fontsize=11)
axes[1, 1].set_title('Vision Transformer - Évolution de l\'Accuracy', fontsize=12, fontweight='bold')
axes[1, 1].legend(fontsize=10)
axes[1, 1].grid(True, alpha=0.3)
axes[1, 1].set_ylim([0, 105])

# ===== 6. Comparaison Accuracy Validation =====
axes[1, 2].plot(epochs_resnet, history_resnet['val_acc'], 'b-', marker='o', label='ResNet18', linewidth=2)
axes[1, 2].plot(epochs_vit, history_vit['val_acc'], 'g-', marker='s', label='Vision Transformer', linewidth=2)
axes[1, 2].set_xlabel('Époque', fontsize=11)
axes[1, 2].set_ylabel('Accuracy Validation (%)', fontsize=11)
axes[1, 2].set_title('Comparaison - Accuracy Validation', fontsize=12, fontweight='bold')
axes[1, 2].legend(fontsize=10)
axes[1, 2].grid(True, alpha=0.3)
axes[1, 2].set_ylim([0, 105])

plt.tight_layout()
plt.show()

# Récupérer les valeurs finales
resnet_train_loss = history_resnet['train_loss'][-1]
resnet_val_loss = history_resnet['val_loss'][-1]
vit_train_loss = history_vit['train_loss'][-1]
vit_val_loss = history_vit['val_loss'][-1]

resnet_train_acc = history_resnet['train_acc'][-1]
resnet_val_acc = history_resnet['val_acc'][-1]
vit_train_acc = history_vit['train_acc'][-1]
vit_val_acc = history_vit['val_acc'][-1]

# Affichage des statistiques détaillées
print("\n" + "="*70)
print("STATISTIQUES COMPARATIVES - LOSS")
print("="*70)
print(f"\nResNet18:")
print(f"  Loss Train initiale: {history_resnet['train_loss'][0]:.4f} → finale: {resnet_train_loss:.4f}")
print(f"  Loss Validation initiale: {history_resnet['val_loss'][0]:.4f} → finale: {resnet_val_loss:.4f}")
print(f"  Amélioration Train: {history_resnet['train_loss'][0] - resnet_train_loss:.4f}")
print(f"  Amélioration Validation: {history_resnet['val_loss'][0] - resnet_val_loss:.4f}")

print(f"\nVision Transformer:")
print(f"  Loss Train initiale: {history_vit['train_loss'][0]:.4f} → finale: {vit_train_loss:.4f}")
print(f"  Loss Validation initiale: {history_vit['val_loss'][0]:.4f} → finale: {vit_val_loss:.4f}")
print(f"  Amélioration Train: {history_vit['train_loss'][0] - vit_train_loss:.4f}")
print(f"  Amélioration Validation: {history_vit['val_loss'][0] - vit_val_loss:.4f}")

print("\n" + "="*70)
print("STATISTIQUES COMPARATIVES - ACCURACY")
print("="*70)
print(f"\nResNet18:")
print(f"  Accuracy Train initiale: {history_resnet['train_acc'][0]:.2f}% → finale: {resnet_train_acc:.2f}%")
print(f"  Accuracy Validation initiale: {history_resnet['val_acc'][0]:.2f}% → finale: {resnet_val_acc:.2f}%")
print(f"  Amélioration Train: {resnet_train_acc - history_resnet['train_acc'][0]:.2f} points")
print(f"  Amélioration Validation: {resnet_val_acc - history_resnet['val_acc'][0]:.2f} points")

print(f"\nVision Transformer:")
print(f"  Accuracy Train initiale: {history_vit['train_acc'][0]:.2f}% → finale: {vit_train_acc:.2f}%")
print(f"  Accuracy Validation initiale: {history_vit['val_acc'][0]:.2f}% → finale: {vit_val_acc:.2f}%")
print(f"  Amélioration Train: {vit_train_acc - history_vit['train_acc'][0]:.2f} points")
print(f"  Amélioration Validation: {vit_val_acc - history_vit['val_acc'][0]:.2f} points")

print(f"\n{'Meilleur modèle (Loss Validation):':.<50} {'ResNet18' if resnet_val_loss < vit_val_loss else 'Vision Transformer'} ({min(resnet_val_loss, vit_val_loss):.4f})")
print(f"{'Meilleur modèle (Accuracy Validation):':.<50} {'ResNet18' if resnet_val_acc > vit_val_acc else 'Vision Transformer'} ({max(resnet_val_acc, vit_val_acc):.2f}%)")

print("\n" + "="*70)
```
